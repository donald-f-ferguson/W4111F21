## Some Insight through Visualization

### Which Were the "Happening Places?"

- The basic structure of the narrative:
    - Season contains multiple episodes.
    - Episodes are sequences of scenes.
    - Scenes occur in a location and have characters.
    
    
- This means that I can count the number of scenes that occur in a location over all episodes and seasons.

```python
# I am going to use some cool packages.
# You will become familiar with the package.
#
import pandas
```

```python
# Run a SQL query that counts how many scenes occur in each (location, sublocation)
# The the result to l_data
#
l_data = %sql SELECT location, subLocation, count(*) as count  FROM gotprocessed.scenes \
                group by location, subLocation having count > 10 order by count desc;
```

```python
# This came back as some kind of Python type/class.
# Specifically a set that contains an SQL query result.
#
type(l_data)
```

- The "cool kids" in the programming track use [Pandas.](https://pandas.pydata.org/)


- Convert the data into Pandas data frames.

```python
l_df = l_data.DataFrame()
l_df
```

```python
l_df = l_df[['location', 'subLocation', 'count']]
```

```python
l_df.head(10)
```

- That's a little hard to understand.


- Human beings are __very good__ at inference from images. Data visualization displays data in formats that allow people to infer information.


- A common visualization is a [heat map](https://en.wikipedia.org/wiki/Heat_map).


- Let's display the information in a heat map.

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
heatmap1_data = pandas.pivot_table(l_df, values='count', 
                     index=['location'], 
                     columns='subLocation')
```

```python
plt.figure(figsize=(20, 10))
sb.heatmap(heatmap1_data, cmap="YlGnBu")

plt.show()
```

- The "happening places" were:
    - (Crownlands, King's Landing)
    - (The North, Winterfell)
    - (The Wall, Castle Black)
